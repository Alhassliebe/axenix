```python
import pandas as pd

delivery = pd.read_excel('ДОСТАВКА.xlsx', header=1)
parking = pd.read_excel('ПАРК.xlsx', header=1)

delivery
```

```python
bins = [0, 500, 1500, float('inf')]
labels = ['<500', '500-1500', '>1500']

for index, row in delivery.iterrows():
    if int(row['Масса груза, кг']) <= 100 and int(row['Путь, км']) <= 20:
        delivery.at[index, 'Прогнозируемая стоимость, руб.'] = row['Путь, км'] * 25
    elif int(row['Масса груза, кг']) <= 500:
        delivery.at[index, 'Прогнозируемая стоимость, руб.'] = row['Путь, км'] * 37
    elif int(row['Масса груза, кг']) <= 1500:
        delivery.at[index, 'Прогнозируемая стоимость, руб.'] = row['Путь, км'] * 51
    else:
        delivery.at[index, 'Прогнозируемая стоимость, руб.'] = row['Путь, км'] * 72

delivery['Масса груза <=100 кг'] = ((delivery['Путь, км'] <= 20) & (delivery['Масса груза, кг'] <= 100)).astype(int)
delivery['Масса груза от 100 до 500 кг'] = ((delivery['Масса груза, кг'] > 100) & (delivery['Масса груза, кг'] <= 500)).astype(int)
delivery['Масса груза от 500 до 1500 кг'] = ((delivery['Масса груза, кг'] > 500) & (delivery['Масса груза, кг'] <= 1500)).astype(int)
delivery['Масса груза >1500 кг'] = (delivery['Масса груза, кг'] > 1500).astype(int)
```

```python
delivery
```

```python
parking = parking.applymap(lambda x: x.lower() if isinstance(x, str) else x)
parking
```

```python
statistics = delivery.groupby('Гос номер').agg(
    mean_cargo_weight=('Масса груза, кг', 'mean'),
    min_cargo_weight=('Масса груза, кг', 'min'),
    max_cargo_weight=('Масса груза, кг', 'max'),
    total_cargo=('Масса груза, кг', 'sum'),
    cargo_weight_under100=('Масса груза <=100 кг', 'sum'),
    cargo_weight_from100to500 =('Масса груза от 100 до 500 кг', 'sum'),
    cargo_weight_from500to1500 = ('Масса груза от 500 до 1500 кг', 'sum'),
    cargo_weight_below1500 = ('Масса груза >1500 кг', 'sum'),
    mean_distance=('Путь, км', 'mean'),
    total_distance=('Путь, км', 'sum'),
    projected_cost=('Прогнозируемая стоимость, руб.', 'sum')
)
```

```python
statistics
```

```python
df = pd.merge(statistics, parking, on='Гос номер')
df['total_cost'] = df['total_distance'] * df['Стоимость пробега / км, руб.']
df
```

```python
for index, row in df.iterrows():
    if int(row['Макс. грузоподъёмность, кг']) == 500:
        df.at[index, '%'] = 100 * row['cargo_weight_under100'] / (row['cargo_weight_under100'] + row['cargo_weight_from100to500'])
    elif int(row['Макс. грузоподъёмность, кг']) == 1500:
        tw = row['cargo_weight_under100'] + row['cargo_weight_from100to500']
        df.at[index, '%'] = 100 * tw / (tw + row['cargo_weight_from500to1500'])
    else:
        tw = row['cargo_weight_under100'] + row['cargo_weight_from100to500'] + row['cargo_weight_from500to1500']
        df.at[index, '%'] = 100 * tw / (tw + row['cargo_weight_below1500'])
df
```

```python
df.to_excel('output.xlsx', index=False)
```
